```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
import catboost as ctb
import shap
```

```python
wine = load_wine(as_frame=True)
df = pd.DataFrame(data= np.c_[wine['data'], wine['target']],
                     columns= wine['feature_names'] + ['target'])
df
```

```python
df = df[df['target'] != 2] 
df.shape
```

```python
y = df['target']
X = df.drop(['target'], axis=1)
X['cat'] = 'some_cat'
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
pool_train = ctb.Pool(X_train, y_train, cat_features = ['cat'])
pool_test = ctb.Pool(X_test, cat_features = ['cat'])

model = ctb.CatBoostRegressor(n_estimators=200,
                       loss_function='RMSE',
                       learning_rate=0.4,
                       depth=3, task_type='CPU',
                       random_state=1,
                       verbose=False)
```

```python
model.fit(pool_train)
y_predict = model.predict(pool_test)
```

```python
explainer = shap.TreeExplainer(model)
shap_values = explainer(pool_test)
```

```python
shap.summary_plot(shap_values, pool_test, plot_type='bar', max_display=100, 
                  show=False, feature_names=wine['feature_names']+['cat'])
```
